```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

class Hood1D:
    """
    z: sites
    n: turns
    p: police presence in each site
    b: initial criminal population
    s: crime level for each site
    beta: arrest rate
    theta: police reactivity
    """
    def __init__(self, z=2, n=10, p=None, b=None,
                 s=None, beta=0.1, theta=0.01):
        self.z = z
        self.n = n
        self.beta = beta
        self.theta = theta
        
        self.p = np.ones(z) if p is None else p
        self.b = np.zeros(z) if b is None else b
        self.s = np.zeros(z) if s is None else s
        
        # initialize grids
        # B: criminals
        self.B = np.zeros((n, z))
        # P: police
        self.P = np.zeros((n, z))
        # A: arrests
        self.A = np.zeros((n, z))
        
        # initial conditions
        self.P[0] = self.p
        self.B[0] = self.b
        
        self.t = 0
    
    """
    transition function:
    For site i, returns array of length z;
    each entry is probability of moving to that
    site from i
    """
    def PI_ij(self, i):
        t = self.t
        p_ = self.P[t]
        x = np.log(1 + np.exp(p_[i] - p_))
        return x / x.sum()
        
    def step(self):
        t = self.t
        # reduce population by arrests
        arrests = self.beta * self.B[t] * self.P[t]
        self.A[t] = arrests   
        b_ = self.B[t] - arrests
        # account for movement
        for i in range(self.z):
            self.B[t+1] += b_[i] * self.PI_ij(i)
        # add from source
        self.B[t+1] += self.s
        # update police numbers
        self.P[t+1] = self.P[t] + self.theta * self.B[t]
        
        self.t += 1
        
    def loop(self):
        for i in range(self.n - 1):
            self.step()
            
    def draw(self):
        a = self.B
        plt.imshow(a, cmap='plasma')
        plt.xticks([])
        plt.yticks([])
        
    def graph(self):
        z = self.z
        n = self.n
        colors = 'brgcmyk'
        fig, ax = plt.subplots()
        for i in range(z):
            ax.plot(range(n), self.B[:, i], '-' + colors[i % len(colors)])

    
```

```python
hood1 = Hood1D()
```

```python
hood1.loop()
```
